```python
#File SimpleCNN

from torch.utils.data import Dataset
import torch.nn as nn
import torch
from google.colab import drive
drive.mount("/content/drive")
import sys
sys.path.append("/content/drive/MyDrive/TestColab")


class SimpleCNNAnimals(nn.Module):
  def __init__(self, num_classes = 10):
    super().__init__()
    self.conv1 = self.make_conv(3, 8)
    self.conv2 = self.make_conv(8, 16)
    self.conv3 = self.make_conv(16, 32)
    self.conv4 = self.make_conv(32, 64)
    self.conv5 = self.make_conv(64, 128)
    self.fc1 = self.make_block_fc(6272,1024)
    self.fc2 = self.make_block_fc(1024,256)
    self.fc3 = nn.Linear(256,num_classes)

  def make_conv(self, input_channel, output_channel):
    return nn.Sequential(
        nn.Conv2d(in_channels= input_channel, out_channels= output_channel, kernel_size= 3, padding = "same", stride = 1),
        nn.BatchNorm2d(output_channel),
        nn.LeakyReLU(),
        nn.Conv2d(output_channel, output_channel, kernel_size = 3, padding = "same", stride = 1),
        nn.BatchNorm2d(output_channel),
        nn.LeakyReLU(),
        nn.MaxPool2d(kernel_size = 2)
    )

  def make_block_fc(self, in_feature, out_feature):
    return nn.Sequential(
        nn.Dropout(0.3),
        nn.Linear(in_feature, out_feature),
        nn.LeakyReLU()
    )

  def forward(self, x):
    x = self.conv1(x)
    x = self.conv2(x)
    x = self.conv3(x)
    x = self.conv4(x)
    x = self.conv5(x)
    x = x.view(x.size(0), -1)
    x = self.fc1(x)
    x = self.fc2(x)
    x = self.fc3(x)
    return x
```
